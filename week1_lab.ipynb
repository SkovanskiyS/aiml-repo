## Problem 1: Python Basics — Clean Summary
Given a list of token counts, compute total, average, min, and max.

## Problem 2: Prompt Engineering — Rule-based Filter
A prompt is *good* if its length is >= 20 chars **and** it contains a `?` or the word `"please"`.


```python
def is_good_prompt(prompt):
    long_enough = len(prompt) >= 20
    has_signal = ("?" in prompt) or ("please" in prompt.lower())
    return long_enough and has_signal


prompts = [
    "Explain transformers?",
    "hi",
    "please summarize this paragraph about GANs",
    "Generate an image of a cat",
]

for p in prompts:
    print(f"{is_good_prompt(p)!s:5}  -> {p!r}")
```

Pair each text length (feature) with its helpfulness label.

## Problem 4: Evaluation Metric — Accuracy
Fraction of predictions equal to the true label.


```python
y_true = [1, 0, 1, 1, 0, 0, 1]
y_pred = [1, 0, 0, 1, 0, 1, 1]


def accuracy(y_true, y_pred):
    correct = sum(t == p for t, p in zip(y_true, y_pred))
    return correct / len(y_true)


print(f"Accuracy: {accuracy(y_true, y_pred):.4f}")
```

## Problem 5: Text Processing — Word Frequency
Lowercase, strip punctuation `.,!?`, and count word frequencies.


```python
text = "GenAI is amazing! GenAI helps students, and GenAI can write code."


def word_frequencies(text):
    cleaned = text.lower()
    for ch in ".,!?":
        cleaned = cleaned.replace(ch, "")
    counts = {}
    for word in cleaned.split():
        counts[word] = counts.get(word, 0) + 1
    return counts


freq = word_frequencies(text)
for word, count in sorted(freq.items(), key=lambda kv: -kv[1]):
    print(f"{word:10} {count}")
```

## Problem 6: Vectors — Dot Product and Cosine Similarity
Implement both from scratch using `math.sqrt`.


```python
import math


def dot(u, v):
    return sum(ui * vi for ui, vi in zip(u, v))


def cosine_similarity(u, v):
    norm_u = math.sqrt(dot(u, u))
    norm_v = math.sqrt(dot(v, v))
    return dot(u, v) / (norm_u * norm_v)


a = [1, 2, 3]
b = [2, 4, 1]

print(f"Dot product:       {dot(a, b)}")
print(f"Cosine similarity: {cosine_similarity(a, b):.4f}")
```

## Problem 7: Softmax — Convert Scores to Probabilities
Return a probability list that sums to 1. Subtract the max for numerical stability.


```python
import math


def softmax(scores):
    shift = max(scores)
    exps = [math.exp(s - shift) for s in scores]
    total = sum(exps)
    return [e / total for e in exps]


scores = [2.0, 1.0, 0.1]
probs = softmax(scores)

print("Probabilities:", [round(p, 4) for p in probs])
print(f"Sum: {sum(probs):.6f}")
```

## Problem 8: Tiny Perceptron Step (1 Update)
Predict `y_hat` then perform a single weight update.


```python
w = [0.2, -0.1]
b = 0.0
x = [1.0, 2.0]
y_true = 1
lr = 0.5


def predict(w, b, x):
    z = sum(wi * xi for wi, xi in zip(w, x)) + b
    return 1 if z >= 0 else 0


y_hat = predict(w, b, x)
error = y_true - y_hat
w = [wi + lr * error * xi for wi, xi in zip(w, x)]
b = b + lr * error

print(f"Prediction y_hat: {y_hat}")
print(f"Updated w:        {w}")
print(f"Updated b:        {b}")
```

## Problem 9: Train a Perceptron on Synthetic Data
10 epochs over a 4-point dataset starting from zero weights.


```python
X = [
    [1, 1],
    [2, 1],
    [2, 3],
    [3, 3],
]
y = [0, 0, 1, 1]

w = [0.0, 0.0]
b = 0.0
lr = 0.1


def predict(w, b, x):
    z = sum(wi * xi for wi, xi in zip(w, x)) + b
    return 1 if z >= 0 else 0


for epoch in range(10):
    for xi, target in zip(X, y):
        y_hat = predict(w, b, xi)
        error = target - y_hat
        w = [wj + lr * error * xij for wj, xij in zip(w, xi)]
        b = b + lr * error

correct = sum(predict(w, b, xi) == target for xi, target in zip(X, y))
accuracy = correct / len(y)

print(f"Final w: {w}")
print(f"Final b: {b}")
print(f"Training accuracy: {accuracy:.4f}")
```

## Problem 10: Mini Language Model — Next-Word Prediction (Bigrams)
Build counts of `(word -> next_word)` and return the most common follower.


```python
sentences = [
    "genai is fun",
    "genai is useful",
    "ai is useful",
    "genai can write code",
    "ai can write code",
]


def build_bigram_counts(sentences):
    counts = {}
    for sentence in sentences:
        words = sentence.split()
        for current, nxt in zip(words, words[1:]):
            counts.setdefault(current, {})
            counts[current][nxt] = counts[current].get(nxt, 0) + 1
    return counts


bigram_counts = build_bigram_counts(sentences)


def predict_next(word):
    if word not in bigram_counts:
        return None
    return max(bigram_counts[word].items(), key=lambda kv: kv[1])[0]


print("Bigram counts:")
for word, followers in bigram_counts.items():
    print(f"  {word!r:8} -> {followers}")

print()
for word in ["genai", "ai", "is", "can", "write"]:
    print(f"predict_next({word!r:8}) = {predict_next(word)!r}")
```
